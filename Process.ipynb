```python
def map_violation_category(description, mapping):
    for key, category in mapping.items():
        if key.lower() in description.lower():
            return category
    return "Other" 
```

If the violation description contains one of the expressions mentionned in the mapping table, this expression should be replaced accordingly.

```python
df['violation_category'] = df['violation_description'].astype(str).apply(lambda x: map_violation_category(x, mapping_violation_cat))
```

```python
df['violation_category'].value_counts()
```

### Merge the less represented violations into Other

In order to have categories with sufficient numbers of occurrences, we cluster all the categories less present than Other in the Other category.

```python
def convert_to_other(row):
    categories = ["No Certificate", "Public Nuisance", "Drugs"]
    
    if row['violation_category'] not in categories:
        return "Other"
    else:
        return row['violation_category']
```

```python
df['violation_category'] = df.apply(convert_to_other, axis=1)
```

We keep only the 2 most representated violations categories. 
We consider all the other ones do not count sufficient number of occurences. 

Having only 2 categories will allow to generalize better.

## Outliers

```python
def min_date(column):
    return df[column].min()
```

```python
def max_date(column):
    return df[column].max()
```

```python
def print_dates(L):
    for element in L:  # 'element' est une chane de caractres correspondant  un nom de colonne
        minimum = min_date(element)
        maximum = max_date(element)
        print(f"The oldest date for {element} is {minimum}")
        print(f"The most recent date for {element} is {maximum}")

L = ["violation_date", "hearing_date", "judgment_date"]
print_dates(L)
```

## Get the Gap between Hearing Date and Judgment Date

```python
# Check the timedelta between the hearing and judgment date
def is_judgement_later(row):
    row['judgment_date'] = pd.to_datetime(row['judgment_date'], format='mixed', errors='coerce')
    row['hearing_date'] = pd.to_datetime(row['hearing_date'], format='mixed', errors='coerce')
    
    if pd.isnull(row['judgment_date']) or pd.isnull(row['hearing_date']):
        return 0
    
    if row['judgment_date'] <= row['hearing_date']:
        return 0
    elif row['judgment_date'].month == row['hearing_date'].month:
        return 1
    elif row['judgment_date'].year == row['hearing_date'].year:
        return 2
    else:
        return 3
```

```python
# Create a new column to see if the judgment date is later than the hearing date
df['is_judgment_later'] = df.apply(is_judgement_later, axis=1)
```
